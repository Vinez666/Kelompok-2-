```python
#import library yang akan digunakan
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
#membaca data
data = pd.read_csv("Keterlibatan Perempuan di Parlemen, 2022.csv")
data.head(553)
```

```python
# creating a rank column and passing the returned rank series 
data["Rank"] = data["Provinsi / Kabupaten / Kota"].rank() 
data.head()
```

```python
# sorting w.r.t name column 
data.sort_values("Provinsi / Kabupaten / Kota", inplace = True) 
data.head(4)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
Kabupaten = ['Aceh', 'Aceh Barat', 'Aceh Barat Daya', 'Aceh Besar']
Nilai = [11.11,8,4,2.86]
ax.pie(Nilai,labels=Kabupaten,autopct='%1.2f%%')
plt.show()
```
